### Importing Libraries

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
#to ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

### Reading Dataset

```python
with open('ncbi_dict.json', 'r') as json_file:
    data = json.load(json_file)
data = pd.DataFrame(data.items(), columns=['ID', 'Sequence'])
```

### Statistics Summary

```python
summary = data.describe()
summary
```

### EDA Univariate Analysis

```python
data['Sequence_Length'] = data['Sequence'].apply(len)
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='Sequence_Length', bins=100, kde=True)
plt.title('Distribution of Sequence Lengths')
plt.xlabel('Sequence Length')
plt.ylabel('Frequency')
plt.show()
```
